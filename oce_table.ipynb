```python
# Assign first row as header

## Access the first row and assign it as column names
new_column_names_df_1 = df_1.iloc[0]
df_1.columns = new_column_names_df_1

## Drop the first row
df_1 = df_1.drop(df_1.index[0])

## Reset the column index
df_1 = df_1.reset_index(drop=True)
```

```python
# Filter by medical speciality

speciality = ('.Onkologie', '.Oncology')
df1_filtered = df_1[df_1['Product: Display Name'].isin(speciality)]
```

```python
# Filter main table (df_2) by values from df1_filtered in column 'Call Name'

df = df_2[df_2['Group'].isin(df1_filtered['Call Name'])]
```

```python
# Select relevant columns from df

df = df.loc[:,['Division', 'Datum/Zeit', 'Group', 'Adresse', 'Kanal', 'Mit Arzt gesprochen', 'weigthed TP', 'CallEQs']]
df.head()
```
