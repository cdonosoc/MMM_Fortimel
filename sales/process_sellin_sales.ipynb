```python
import pandas as pd
```

```python
df = pd.read_excel('Fortimel Sales (Sell-in) (CAF) aggregated.xlsx')
```

```python
df
```

```python
df['week_date_temp'] = df['(CAL) Week To date'].astype(str).str[:4] + '-' + df['(CAL) Week To date'].astype(str).str[4:6] + '-' + df['(CAL) Week To date'].astype(str).str[6:8]
```

```python
df
```

```python
df['week_end_date'] = pd.to_datetime(df['week_date_temp'])
```

```python
df
```

```python
df.info()
```

```python
df = df[df['week_end_date'] > '2021-01-31']
df
```

```python
df = df[df['week_end_date'] < '2023-06-19']
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
sales = df[['week_end_date', 'Invoiced Sales (CAF)']]
```

```python
sales
```

```python
sales.to_csv('sales.csv')
```

```python
costs = pd.read_excel('oce_table_costs.xlsx')
```

```python
costs
```

```python
final_df = costs.merge(sales, left_on='date', right_on='week_end_date', how='inner')
```

```python
final_df
```

```python
final_df.to_excel('final_df.xlsx')
```
